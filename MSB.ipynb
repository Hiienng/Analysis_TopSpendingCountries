```python
from bs4 import BeautifulSoup
import requests
```

```python
MSB = 'https://www.msb.com.vn/vi/w/ca-nhan/tiet-kiem/tiet-kiem-co-ky-han-truc-tuyen'

MSB_page = requests.get (MSB)

MSB_soup = BeautifulSoup(MSB_page.text, 'html') 
```

```python
print (MSB_soup)
```

```python
MSB_table = MSB_soup.find('table')
```

```python
print (MSB_table)
```

```python
MSB_title_raw = MSB_table.find('thead')
MSB_title = MSB_title_raw.find_all ('tr')
```

```python
print (MSB_title)
```

```python
#MSB_table_title = [title.text.strip() for title in MSB_title]
MSB_table_title = ['Đơn vị tính', 'Hợp đồng tiền gửi', 'Trả lãi ngay - GTQ', 'Trả lãi ngay - GTT', 'Định kỳ sinh lời - GTQ'
                  , 'Định kỳ sinh lời - GTT', 'Rút gốc từng phần - GTQ', 'Rút gốc từng phần - GTT', 'Lãi suất cao - GTQ'
                  , 'Lãi suất cao- GTT' , 'Măng Non', 'Ong vàng - GTQ', 'Ong Vàng - GTT']

print (MSB_table_title)
```

```python
import pandas as pd 
```

```python
df = pd.DataFrame (columns = MSB_table_title)

df
```

```python
body = MSB_table.find('tbody')
column_data = body.find_all ('tr')
```

```python
for row in column_data:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print (individual_row_data)
    
    length = len(df)
    df.loc[length] = individual_row_data
```
