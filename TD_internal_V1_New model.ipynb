```python

# Tạo DataFrame với mapping chuẩn.
months = []
for year in range(2023, 2025):
    for month in range(1, 13):
        month_str = f"{year}-{month:02}"
        months.append(month_str)

Period = np.array(months, dtype='object')

FCST_RESULTS_TD_ALL = pd.DataFrame(columns=['Vintages', 'Financial_Items', 'Type', 'Term_CM', 'Currency_2', 'Period', 'Indicators', 'Month_ID', 'Month_end', 'Value'])

Term_CM = df_movement_term['Term_CM'].unique()
Currency_2 = df_movement_term['Currency_2'].unique()
Indicator = ['1. Opening Balance',  
                          '3. Closing Balance' , 
                          '2.1 Not-roll: New sales',
                          '2.1 Not-roll: Schedule on system',
                          '2.1 Not-roll: Premature', 
                          '2.2 Autoroll: New sales',
                          '2.2 Autoroll: Roll',
                          '2.2 Autoroll: Premature', 
                          '2.3 Other',
                        '%Premature Not-roll/EOP Not-roll',
                        '%Premature Autoroll/EOP Autoroll',
                        '%Roll_Autoroll',
                        '%NewNot-roll',
                        '%New Autoroll',
                        '%Schedule/Not-roll'
                        ]

Term_CM, Currency_2, Period, Indicator = np.meshgrid(Term_CM, Currency_2, Period, Indicator)
Term_CM = Term_CM.flatten()
Currency_2 = Currency_2.flatten()
Period = Period.flatten()
Indicator = Indicator.flatten()


df = pd.DataFrame({'Vintages': 'ALL',
                   'Financial_items' : 'EOP',
                   'Type': 'ALL',
                   'Term_CM': Term_CM, 
                   'Currency_2': Currency_2,
                   'PERIOD': Period,
                   'INDICATORS': Indicator
                   })
df
```

```python
# Tạo hàm tùy chỉnh để lấy giá trị từ df_movement_term
def EOP_Opening_Balance_AndOther(row):
    '''
    Nghĩa: 
    + Nếu dòng là OB thì giá trị trả về dòng là tổng các EOP có nhóm tương ứng
    + Nếu không
    '''
    if row['INDICATORS'] == '1. Opening Balance':
        term_cm = row['Term_CM']
        currency_2 = row['Currency_2']
        period = row['PERIOD']

        value = df_movement_term.loc[
            (df_movement_term['Term_CM'] == term_cm) 
            & (df_movement_term['Currency_2'] == currency_2)
            & (df_movement_term['PERIOD'] == period)
            , 'EOP_LM'
        ].values

        # Trả về giá trị hoặc 0 nếu không tìm thấy
        if len(value) > 0:
            return value.sum()
        else:
            return 0
    else:  # Added 'else' to fix indentation
        term_cm = row['Term_CM']
        currency_2 = row['Currency_2']
        period = row['PERIOD']
        indicator = mapping_indicator.get(row['INDICATORS'], None)

        value = df_movement_term.loc[
            (df_movement_term['Term_CM'] == term_cm) 
            & (df_movement_term['Currency_2'] == currency_2)
            & (df_movement_term['PERIOD'] == period)
            & (df_movement_term['MOVEMENT_TYPE'] == indicator)
            , 'NET'
        ].values

        # Trả về giá trị hoặc 0 nếu không tìm thấy
        if len(value) > 0:
            return value.sum()
        else:
            return 0

def EOP_Closing_Balance(row):
    term_cm = row['Term_CM']
    currency_2 = row['Currency_2']
    period = row['PERIOD']

    # Locate rows in df_movement_term for the current 'Term_CM', 'Currency_2', and 'PERIOD'
    matching_rows = df_movement_term.loc[
        (df_movement_term['Term_CM'] == term_cm) 
        & (df_movement_term['Currency_2'] == currency_2)
        & (df_movement_term['PERIOD'] == period)
    ]

    # Sum the 'EOP_LM' values for the matching rows (Opening Balance)
    opening_balance = matching_rows['EOP_LM'].sum()

    # Sum the 'NET' values for other indicators in df_movement_term
    other_indicators_sum = matching_rows['NET'].sum()

    # Calculate the closing balance as the sum of opening balance and other indicators
    closing_balance = opening_balance + other_indicators_sum

    return closing_balance

# Định nghĩa mapping_indicator
mapping_indicator = {'2.1 Not-roll: New sales': "1NEW_NOTROLL",
                    '2.1 Not-roll: Schedule on system': "2SCHE_NOTROLL",
                    '2.1 Not-roll: Premature': "3PREMATURE_NOTROLL", 
                    '2.2 Autoroll: New sales': "5NEW_AUTOROLL",
                    '2.2 Autoroll: Roll': "6ROLL_AUTOROLL",
                    '2.2 Autoroll: Premature': "7PREMATURE_AUTOROLL", 
                    '2.3 Other': "9OTHERS"
}

List_af_OB = { '1. Opening Balance',
            '2.1 Not-roll: New sales',
            '2.1 Not-roll: Schedule on system',
            '2.1 Not-roll: Premature', 
            '2.2 Autoroll: New sales',
            '2.2 Autoroll: Roll',
            '2.2 Autoroll: Premature', 
            '2.3 Other',
}

# Apply the opening balance function to rows where INDICATORS is not '1. Closing Balance'
df['VALUE'] = df.apply(lambda row: EOP_Opening_Balance_AndOther(row) if row['INDICATORS'] in List_af_OB else 0, axis=1)

# Apply the closing balance function to rows where INDICATORS is '1. Closing Balance'
df['VALUE'] += df.apply(lambda row: EOP_Closing_Balance(row) if row['INDICATORS'] == '3. Closing Balance' else 0, axis=1)

# In ra DataFrame
df
```

```python
'''  
Công thức:

Rate_NotRoll_Pre = % EOP of Premature in Not Roll
Rate_NotRoll_NewSales = % EOP of New Sales in Not Roll
Rate_NotRoll_Sche = % EOP of Schedule on system in Not Roll

Rate_Roll_Pre = % EOP of NPremature in Roll
Rate_Roll_NewSales = % EOP of New Sales in Roll
Rate_Roll_Auto = % EOP of Auto in Roll
'''
from datetime import datetime, timedelta

def EOP_Rate(row):
    term_cm = row['Term_CM']
    currency_2 = row['Currency_2']
    period = datetime.strptime(row['PERIOD'], '%Y-%m')
    indicator = mapping_indicator.get(row['INDICATORS'], None)

    # Locate rows in df_movement_term for the following rules
    matching_rows_Lv3 = df_movement_term.loc[
        (df_movement_term['Term_CM'] == term_cm) 
        & (df_movement_term['Currency_2'] == currency_2)
        & (df_movement_term['PERIOD'] == period)
        & (df_movement_term['MOVEMENT_TYPE'] == indicator)
    ]
    previous_period = period - timedelta(months=1)
    # Locate rows in df_movement_term for the following rules
    matching_rows_Lv2 = df_movement_term.loc[
        (df_movement_term['Term_CM'] == term_cm) 
        & (df_movement_term['Currency_2'] == currency_2)
        & (df_movement_term['PERIOD'] == previous_period)
    ]
    NotRoll_Pre = matching_rows_Lv3['NET'].sum()
    NotRoll = matching_rows_Lv2['EOP_CM'].sum()
    Rate_NotRoll_Pre = NotRoll_Pre/NotRoll

    return Rate_NotRoll_Pre

List_Rate = {'%Premature Not-roll/EOP Not-roll',
            '%Premature Autoroll/EOP Autoroll',
            '%Roll_Autoroll',
            '%NewNot-roll',
            '%New Autoroll',
            '%Schedule/Not-roll'
            }

df['VALUE'] += df.apply(lambda row: EOP_Rate(row) if row['INDICATORS'] in List_Rate else 0, axis=1)
```
