```python
from bs4 import BeautifulSoup
import requests
import selenium
```

```python
#import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
#driver download link:
#(https://googlechromelabs.github.io/chrome-for-testing/known-good-versions-with-downloads.json) - based on chrome ver
driver = webdriver.Chrome(executable_path = "C:\chromedriver.exe")

#wait for everything to load - change based on internet speed
driver.implicitly_wait (10)
#maximize window
driver.maximize_window()
driver.get ('https://www.vietcombank.com.vn/vi-VN/KHCN/Cong-cu-Tien-ich/KHCN---Lai-suat')
#click on element
TKTQ = driver.find_element(by=By.XPATH, value="//*[@id='content']/div[5]/div/div[1]/div/div[1]/div[1]/button[1]").click()
```

```python
VCB_html = driver.page_source

VCB_soup = BeautifulSoup(VCB_html, 'html5lib') 
```

```python
print (VCB_soup)
```

```python
VCB_table = VCB_soup.find ('div', class_= "table-wrapper")
```

```python
print (VCB_table)
```

```python
VCB_title_raw = VCB_table.find('thead')
VCB_title = VCB_title_raw.find_all ('span', class_="d-desktop-none")
```

```python
print (VCB_title)
```

```python
VCB_table_title = [title.text.strip() for title in VCB_title]


print (VCB_table_title)
```

```python
import pandas as pd 
```

```python
df = pd.DataFrame (columns = VCB_table_title)

df
```

```python
body = VCB_table.find('tbody')
column_data = body.find_all ('tr')
```

```python
for row in column_data:
    row_data = row.find_all('td')
    individual_row_data = [data.text.strip() for data in row_data]
    print (individual_row_data)
    
    length = len(df)
    df.loc[length] = individual_row_data
```
